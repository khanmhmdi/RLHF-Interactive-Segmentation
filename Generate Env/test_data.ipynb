```python
import pickle
import matplotlib.pyplot as plt

path = '/home/kasra/Desktop/New Folder/PosEstimation/Generate Env/env_data/00146_t1_112_0_0.pkl'
with open(path, 'rb') as f:
    data = pickle.load(f)

denormalized_image = data[1]
padded_mask = data[5]
label = data[7]
low_res_mask = data[8]
print(denormalized_image.shape)
print(padded_mask.shape)
print(label.shape)
print(low_res_mask.shape)

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].imshow(denormalized_image[0], cmap='gray')
axes[0].set_title('Denormalized Image')
axes[0].axis('off')

axes[1].imshow(padded_mask, cmap='gray')
axes[1].set_title('Padded Mask')
axes[1].axis('off')

axes[2].imshow(label, cmap='gray')
axes[2].set_title('Label')
axes[2].axis('off')

axes[3].imshow(low_res_mask, cmap='gray')
axes[3].set_title('Low Res Mask')
axes[3].axis('off')

plt.tight_layout()
plt.show()
```
